```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/train.csv')
df.shape
```

Eliminación de nans

```python
cols_to_drop = df.columns[df.isna().mean() > 0.6]
df = df.drop(cols_to_drop, axis=1)
df.shape
```

```python
null_counts = df.select_dtypes(include='number').isna().sum()
null_counts[null_counts > 0]
```

```python
for col in ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']:
    median = df[col].median(numeric_only=True)
    df[col] = df[col].fillna(median)
    
```

```python
df[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].isna().sum()
```

```python
null_categorical = df.select_dtypes(include='object').isna().sum()
null_categorical[null_categorical > 0]
```

```python
for col in null_categorical[null_categorical > 0].index.drop('Electrical').to_list():
    df[col] = df[col].fillna('None')

mode_elec = df['Electrical'].mode()[0]
df['Electrical'] = df['Electrical'].fillna(mode_elec)
    
```

```python
df.isna().sum().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df['SalePrice'] = np.log(df['SalePrice'])
```

```python
prices_log = df['SalePrice']
histo_log = sns.histplot(data=df, x= prices_log)
```

```python
df.isna().sum().sum()
```

```python
df.to_csv('../data/processed/train_clean.csv', index=False)
```
